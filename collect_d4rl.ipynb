```python
import h5py
import numpy as np

# 打开 .hdf5 文件
with h5py.File('/mnt/nfs/work/albertliu/.d4rl/datasets/hopper_expert-v2.hdf5', 'r') as f:
    # 假设 'actions' 和 'observations' 是数据集的名称
    actions = f['actions'][:]  # 读取 actions 数据集
    observations = f['observations'][:]  # 读取 observations 数据集

    # 假设每条轨迹的 timestep 为 1000，你可以选择前 100 条轨迹
    num_trajs = 15  # 你只需要 100 条轨迹
    timesteps_per_traj = 1000  # 每条轨迹有 1000 个时间步长

    actions_subset = actions[:num_trajs * timesteps_per_traj]  # 获取前 100 条轨迹的 actions
    observations_subset = observations[:num_trajs * timesteps_per_traj]  # 获取前 100 条轨迹的 observations

    # 将 actions 和 observations 分别保存为 .npy 文件
    np.save('actions_subset.npy', actions_subset)
    np.save('observations_subset.npy', observations_subset)

    print(f"Actions subset shape: {actions_subset.shape}")
    print(f"Observations subset shape: {observations_subset.shape}")
    print("Subset data successfully saved to actions_subset.npy and observations_subset.npy")
```
