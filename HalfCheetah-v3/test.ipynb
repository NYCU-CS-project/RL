```python
import gymnasium as gym
import imageio
env = gym.make('ALE/SpaceInvaders', render_mode='rgb_array')
state = env.reset()
done = False
total_reward = 0
frames = []
max_steps = 1000
max_single_reward = 0
while not done and max_steps > 0:
    max_steps -= 1
    action = env.action_space.sample()
    state, reward,_, done, _ = env.step(action)
    total_reward += reward
    if reward > max_single_reward:
        max_single_reward = reward
    frame = env.render()
    frames.append(frame)
print('Total reward:', total_reward)
print('Max single reward:', max_single_reward)
imageio.mimsave('./ms_pacman_random.gif', frames, duration=1/30)
```
